```python
import os
import numpy as np
import pandas as pd

# 設定 data_path
dir_data = 'd:/data/'

f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
```

```python
app_train.describe()
```

## (1)資料的大小

```python
s = app_train.size
print('size of application_train: %s' % (s))
(r,c) = app_train.shape
print('rows of application_train: %s' % (r))
print('columns of application_train: %s' % (c))
```
